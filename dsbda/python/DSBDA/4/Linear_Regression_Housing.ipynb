# Simple / Multiple Linear Regression on Housing Data

This notebook loads the housing CSV, cleans missing values, trains a Linear Regression model, and evaluates it using MSE, MAE, and RMSE.

```
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```
# Load dataset
df = pd.read_csv("housing_data(2).csv")

# Display first 5 rows
df.head()
```

```
# Check structure
df.info()

# Statistical summary
df.describe()
```

```
# Check missing values
df.isnull().sum()
```

```
# Remove missing values
df = df.dropna()

# Confirm cleaning
df.isnull().sum()
```

```
# X = input features, y = target (MEDV = price)
X = df.drop("MEDV", axis=1)
y = df["MEDV"]
```

```
# Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42
)
```

```
# Create Linear Regression model
model = LinearRegression()
```

```
# Train model
model.fit(X_train, y_train)
```

```
# Display coefficients
print("Coefficients:", model.coef_)
```

```
# Predict on test data
y_pred = model.predict(X_test)

# Show predictions
print("Predicted values:", y_pred[:10])
print("Actual values:", y_test.values[:10])
```

```
# Scatter plot (Actual vs Predicted)
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted")
plt.show()
```

```
# Calculate errors
mse = metrics.mean_squared_error(y_test, y_pred)
mae = metrics.mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
```

## Note

This is **multiple linear regression** because it uses many input columns to predict `MEDV`.
